```python
import pandas as pd
data={
    "Employee_ID":[101,102,103,104,105,106,107,108,109,110],
    "Name":["Amith","Priya","Rahul","Sneha","Karan","Divya","Arjun","Pooja","Vikram","Neha"],
    "Department":["Hr","IT","Finance","IT","Marketing","Hr","Finance","IT","Marketing","HR"],
    "Salary":[45000,60000,75000,50000,65000,48000,80000,52000,70000,55000]

}
df = pd.DataFrame(data)
```

```python
print("original DataFrame:\n")
print(df)
```

```python
#02
print("\nFirst 5 Rows: \n",df.head())
print("\n Shape: \n",df.shape)
print("\n Column Names:",df.columns.tolist())
```

```python
#03
print("\nHighest Salary:",df["Salary"].max())
print("\nLowest Salary:",df["Salary"].min())
print("\nAverage Salary:",df["Salary"].mean())
print("\Total Employee Count:",df.shape[0])
```

```python
#04
print("\n Employee earning more than average salary:\n")
print(df[df["Salary"]>df["Salary"].mean()])
```

```python
#05
print("\n Employee earning more than 50k salary:\n")
print(df[df["Salary"]>50000])
```

```python
#06
print("/nEmployee from IT Department:\n")
print(df[df["Department"]=="IT"])
```

```python
#07
print("\n Employee sorted by salary in asending: \n")
print(df.sort_values(by="Salary"))
```

```python
print("\n Employee sorted by salary in asending: \n")
print(df.sort_values(by="Salary",ascending=False))
```

```python
#08
print("\n Employee sorted by salary in asending: \n")
print(df.sort_values(by="Salary",ascending=False).head(3))
```

```python
df["Bonus"]=df["Salary"]*0.10
print(df)
```

```python
df["Total_pay"]=df["Salary"]+df["Bonus"]
print("\n After Adding Bonus and Total_pay:\n")
print(df)
```

```python
df.loc[0,'Marks'] = None
df.loc[3,'Marks'] = None
print(df.head())
```

```python
#15
avg_marks = df['Marks'].mean()
df['Marks'].fillna(avg_marks,inplace=True)
print(df.head())
```

```python
#16
print("Duplicate Rows:" ,df.duplicated().sum())
df=df.drop_duplicates()
print("After Removing Duplicates:",df.duplicated().sum())
```

```python
df['Result']=df['Marks'].apply(lambda x:'pass' if x>=40 else'Fail')
print(df.head())
```

```python
df.rename(columns={'Marks': 'Final_marks'}, inplace=True)
print(df.head())
```

```python
df.drop(columns=['Subject'],inplace=True)
print(df.head())
```

```python
df.to_csv('final_students.csv',index=False)
```
